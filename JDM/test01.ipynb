```python
df.rename(columns=({'MDVP:Fo(Hz)':'avg_fre', 'MDVP:Fhi(Hz)':'max_fre', 'MDVP:Flo(Hz)':'min_fre', 'MDVP:Jitter(%)':'var_fre1',
       'MDVP:Jitter(Abs)':'var_fre2', 'MDVP:RAP':'var_fre3', 'MDVP:PPQ':'var_fre4', 'Jitter:DDP':'var_fre5',
       'MDVP:Shimmer':'var_amp1', 'MDVP:Shimmer(dB)':'var_amp2', 'Shimmer:APQ3':'var_amp3', 'Shimmer:APQ5':'var_amp4',
       'MDVP:APQ':'var_amp5', 'Shimmer:DDA':'var_amp6'}),inplace=True)
df
```

avg_fre : 평균 주파수(음성의 높낮이)
var_fre1 : 주파수 변동성(목소리의 불안정성이나 불규칙성 평가)
var_fre2 : 절대 주파수 변동(백분율이 아닌 실제 주파수의 변화량)
var_fre3 : 주파수의 RAP (상대적인 진폭 변동 측정하여 음성의 불규칙성 나타냄)
var_fre4 : 주파수의 PPQ(음성의 피치 변동을 나타내며 음성의 불안정성과 관련된 지표)
var_fre5 : 주파수의 DDP(음성의 피치가 얼마나 불규칙하게 변하는지 측정)
var_amp1 : 진폭의 변동성(목소리의 일관성을 나타내는 데 사용)
var_amp2 : 데시벨 단위의 진폭 변동성
var_amp3 : 진폭의 APQ3 (진폭의 변화량을 나타내는 지표로 음성의 불규칙성 평가)
var_amp4 : 진폭의 APQ5 (음성의 진폭 변동)
var_amp5 : 진폭의 APQ (진폭 변화)
var_amp6 : DDA (진폭에서의 불규칙적인 변동을 나타내며 음성의 불안정성 측정)
NHR : 소리의 고조파 성분과 잡음 성분의 비율
HNR : 고조파 성분과 잡음 성분의 비율
status : 1:파킨슨병, 0:건강
RPDE : 신호의 상대 엔트로피, 신호의 복잡성이나 불규칙성으 나타냄
DFA : 감소된 변동성 분석
spread1 : 신호의 스프레드 또는 확산 지표
spread2 : 두 번째 스프레드
D2 : 상관 차원
PPE : 음성의 주기적 성분의 엔트로피

Determining Depentend & Independent Variables

```python
# get features and labels

x=df.loc[:,df.columns!='status'].values[:,1:] #머신러닝 모델에 데이터를 넣기 위해 numpy 배열 형태
x1=df.loc[:,df.columns!='status'] #데이터프레임 형태
y=df.loc[:,'status'].values
y1=df.loc[:,'status']
```

```python
df=df[df.max_fre<=300]
df=df[df.var_fre1<=0.02]
df=df[df.var_fre2<=0.0001]
df=df[df.var_fre3<=0.01]
df=df[df.var_fre4<=0.01]
df=df[df.var_fre5<=0.02]
df=df[df.var_amp1<=0.10]
df=df[df.var_amp2<=1.0]
df=df[df.var_amp3<=0.04]
df=df[df.var_amp4<=0.050]
df=df[df.var_amp5<=0.075]
df=df[df.var_amp6<=0.125]
df=df[df.NHR<=0.15]
```

Correlation Matrix

```python
correl=x1.corr() # 상관관계 분석
plt.figure(figsize=(20,20))
sns.heatmap(correl,annot=True,cmap='OrRd')
plt.show()
```

```python
#Scale the features to between -1 and 1
scaler=MinMaxScaler((-1,1)) #정규화 도구
x1=scaler.fit_transform(x) #x데이터 정규화=-1과1사이 값(데이터의 범위가 너무 커 최적화가 안 되기 때문)
y1=y
```

```python
#Split the dataset

xtrain,xtest,ytrain,ytest=train_test_split(x1, y1, test_size=0.2) #전체 데이터의 20%만 테스트, 나머지는 훈련, split : 데이터를 무작위로 분리, 과적합 방지
```

```python
# Train the model
from xgboost import XGBClassifier


model=XGBClassifier()
model.fit(xtrain,ytrain)
predict=model.predict(xtest)
```

```python
print(accuracy_score(ytest,predict)*100) #실제값 모델 예측 값 비교
```

Implementing Confusion Matrix

```python
from sklearn.metrics import confusion_matrix #혼동 행렬
cm=confusion_matrix(ytest,predict)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Reds") #cm = 혼동행렬 , annot=True 각 셀에 숫자 표시
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```
