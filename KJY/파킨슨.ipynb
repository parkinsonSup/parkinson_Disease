```
# 'name' 컬럼은 분석에 불필요하므로 제거
if 'name' in df.columns:
    df = df.drop('name', axis=1)

# 독립 변수(X)와 종속 변수(y) 분리
X = df.drop('status', axis=1)
y = df['status']

# 학습 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
print("\n--- 상관 분석 결과 ---") #필터 방식
correlations = X_train.corrwith(y_train) #DataFrame의 각 열과 y_train 간의 피어슨 상관 계수 계산
significant_corr_features = correlations[abs(correlations) > 0.4].index.tolist() # 적절한 임계값 설정 여기서는 0.4로 해봄
print(correlations.sort_values(ascending=False))
print(f"\n상관관계가 높은 변수 (절대값 > 0.4): {significant_corr_features}")

# 상관관계 시각화 (히트맵)
plt.figure(figsize=(8, 6))
sns.heatmap(X_train.corr(), annot=False, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()

plt.figure(figsize=(8, 6))
correlations.sort_values().plot(kind='barh')
plt.title('Correlation with Status')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Feature')
plt.show()
```

```
print("\n--- Random Forest 특성 중요도 ---") #임베디드 방법
rf_model = RandomForestClassifier(n_estimators=100, random_state=42) #트리 100개
rf_model.fit(X_train, y_train)
importances = rf_model.feature_importances_
feature_importances = pd.Series(importances, index=X_train.columns)
sorted_importances = feature_importances.sort_values(ascending=False)

print(sorted_importances)

significant_rf_features = sorted_importances[sorted_importances > 0.04].index.tolist() # 적절한 임계값 설정 여기서는 0.04
print(f"\nRandom Forest 중요 변수 (importance > 0.04): {significant_rf_features}")

# 특성 중요도 시각화
plt.figure(figsize=(10, 8))
sns.barplot(x=sorted_importances, y=sorted_importances.index)
plt.title('Random Forest Feature Importance')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# 데이터 표준화 (PCA는 스케일링에 민감함)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA 수행 (모든 주성분 유지)
n_components = X.shape[1]
pca = PCA(n_components=n_components)
pca.fit(X_scaled)

# 설명된 분산 비율 확인 (Optional)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)
print("설명된 분산 비율:", explained_variance_ratio[:6]) # 상위 6개 주성분
print("누적 설명된 분산 비율:", cumulative_variance_ratio[:6]) # 상위 6개 주성분

# 주성분-원변수 관계 (Loading Vectors) 확인
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(n_components)], index=X.columns)
print("\n주성분 부하량 (Loading Vectors):\n", loadings.iloc[:, :6]) # 상위 6개 주성분에 대한 로딩 값

# PCR 수행 (선택된 주성분으로 로지스틱 회귀)
n_components_selected = 6
pca_selected = PCA(n_components=n_components_selected)
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
X_train_pca = pca_selected.fit_transform(X_train_scaled)
X_test_pca = pca_selected.transform(X_test_scaled)

logistic_model_pcr = LogisticRegression(random_state=42)
logistic_model_pcr.fit(X_train_pca, y_train)
y_pred_pcr = logistic_model_pcr.predict(X_test_pca)
accuracy_pcr = accuracy_score(y_test, y_pred_pcr)
print(f"\nPCR 기반 로지스틱 회귀 정확도 (주성분 {n_components_selected}개 사용): {accuracy_pcr:.4f}")

# 중요 변수 선정 (loading 값 기준으로)
threshold = 0.5 # 임계값 조절 필요
important_features_pca = set()
print("\n각 주성분에 대한 중요 변수 (loading 절댓값 > {}):".format(threshold))
for i in range(n_components_selected):
    pc_name = f'PC{i+1}'
    important_features_pc = loadings[abs(loadings[pc_name]) > threshold].index.tolist()
    print(f"{pc_name}: {important_features_pc}")
    important_features_pca.update(important_features_pc)

print(f"\n종합적으로 고려할 PCA 기반 중요 변수 (loading 절댓값 > {threshold}): {list(important_features_pca)}")
```

```
def initilization_of_population(size,n_feat): #유전자 알고리즘
    population = []
    for i in range(size):
        chromosome = np.ones(n_feat,dtype=np.bool)
        chromosome[:int(0.3*n_feat)]=False
        np.random.shuffle(chromosome)
        population.append(chromosome)
    return population


def fitness_score(population):
    scores = []
    for chromosome in population:
        logmodel.fit(X_train.iloc[:,chromosome],Y_train)
        predictions = logmodel.predict(X_test.iloc[:,chromosome])
        scores.append(accuracy_score(Y_test,predictions))
    scores, population = np.array(scores), np.array(population)
    inds = np.argsort(scores)
    return list(scores[inds][::-1]), list(population[inds,:][::-1])


def selection(pop_after_fit,n_parents):
    population_nextgen = []
    for i in range(n_parents):
        population_nextgen.append(pop_after_fit[i])
    return population_nextgen
```

```
def crossover(pop_after_sel):
    pop_nextgen = pop_after_sel
    for i in range(0,len(pop_after_sel),2):
        new_par = []
        child_1 , child_2 = pop_nextgen[i] , pop_nextgen[i+1]
        new_par = np.concatenate((child_1[:len(child_1)//2],child_2[len(child_1)//2:]))
        pop_nextgen.append(new_par)
    return pop_nextgen



def mutation(pop_after_cross,mutation_rate,n_feat):
    mutation_range = int(mutation_rate*n_feat)
    pop_next_gen = []
    for n in range(0,len(pop_after_cross)):
        chromo = pop_after_cross[n]
        rand_posi = []
        for i in range(0,mutation_range):
            pos = randint(0,n_feat-1)
            rand_posi.append(pos)
        for j in rand_posi:
            chromo[j] = not chromo[j]
        pop_next_gen.append(chromo)
    return pop_next_gen

def generations(df,label,size,n_feat,n_parents,mutation_rate,n_gen,X_train,
                                   X_test, Y_train, Y_test):
    best_chromo= []
    best_score= []
    population_nextgen=initilization_of_population(size,n_feat)
    for i in range(n_gen):
        scores, pop_after_fit = fitness_score(population_nextgen)
        print('Best score in generation',i+1,':',scores[:1])  #2
        pop_after_sel = selection(pop_after_fit,n_parents)
        pop_after_cross = crossover(pop_after_sel)
        population_nextgen = mutation(pop_after_cross,mutation_rate,n_feat)
        best_chromo.append(pop_after_fit[0])
        best_score.append(scores[0])
    return best_chromo,best_score
```
